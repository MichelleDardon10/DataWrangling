```python
pysqldf = lambda q: sqldf(q,globals())
```

```python
pysqldf(query2)
```

```python
query1 = """
SELECT *
FROM heroes_information
ORDER BY name
LIMIT 5
"""
```

```python
pysqldf(query1)
```

```python
query2 = """
SELECT COUNT(DISTINCT Publisher)
FROM heroes_information
"""
```

```python
pysqldf(query2)
```

```python
query3 = """
SELECT COUNT(*)
FROM heroes_information
WHERE Height > 200
"""
pysqldf(query3)
```

```python
query4 = """
SELECT COUNT(*)
FROM heroes_information
WHERE Height > 200 AND Race = "Human"
"""
pysqldf(query4)
```

```python
query5 = """
SELECT COUNT(*)
FROM heroes_information
WHERE Weight BETWEEN 100 AND 200
"""
pysqldf(query5)
```

```python
query6 = """
SELECT COUNT(*)
FROM heroes_information
WHERE `Eye color` = "red" OR `Eye color`= "blue"
"""

pysqldf(query6)
```

```python
query7 = """
SELECT COUNT(*)
FROM heroes_information
WHERE (race = "Human" OR race = "Mutant") AND `Hair color`= "Green"
    OR (race = "Vampire" AND `Hair color` = "Black")
"""

pysqldf(query7)
```

```python
query8 = """
SELECT *
FROM heroes_information
ORDER BY race DESC
LIMIT 5
"""

pysqldf(query8)
```

```python
query9 = """
SELECT COUNT(*)
FROM heroes_information
GROUP BY gender
"""

pysqldf(query9)
```

```python
query10 = """
SELECT COUNT(Publisher)
FROM heroes_information
GROUP BY publisher
HAVING COUNT(*) > 15
"""

pysqldf(query10)
```
